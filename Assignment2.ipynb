1. What are the two values of the Boolean data type? How do you write them?
Boolean type is one of the built-in data types provided by Python, which represents one of the two values i.e. True or False. Generally, it is used to represent the truth values of the expressions. 
For example, 1==1 is True whereas 2==3 is False.
